### To check whether given sides form a triangle or not check below
Sum of 2 sides > Third side

```python
class Triangle:
    
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        
    def isTriangle(self):
        a = self.side1
        b = self.side2
        c = self.side3
        if (a+b)>c and (b+c)>a and (c+a)>b:            
            s = True
        else:            
            s = False
        return s
    
    def Perimeter(self):
        if self.isTriangle()==True:
            p = self.side1 + self.side2 + self.side3
        else:
            print(f'The given sides {self.side1}, {self.side2} and {self.side3} DO NOT form Triangle')
            p = None
        return p
    
    def Area(self):
        a = self.side1
        b = self.side2
        c = self.side3
        if self.isTriangle()==True:
            s = (a+b+c)/2
            ar = (s*(s-a)*(s-b)*(s-c))**(1/2)
        else:
            ar = None
            print(f'The given sides {a}, {b} and {c} DO NOT form a Triangle') 
        return ar
        
```
